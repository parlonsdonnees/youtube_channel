# 3. Athlete Event Preparation - Joining the Tables

```python
def extract_season(edition):
    if 'Summer' in edition:
        return 'Summer'
    elif 'Winter' in edition:
        return 'Winter'
    else:
        return 'na'
olympic_games_df['season'] = olympic_games_df['edition'].apply(extract_season)
olympic_games_df.head()
```

```python
agg_athlete_event_results_df = pd.merge(
    olympic_athlete_event_results_df.filter(items=['athlete_id', 'result_id', 'pos', 'medal', 'isTeamSport']),
    olympic_athlete_bio_df,
    how='left',
    on='athlete_id')

agg_athlete_event_results_df = pd.merge(
    agg_athlete_event_results_df,
    olympic_results_df,
    how='left',
    on='result_id')

agg_athlete_event_results_df = pd.merge(
    agg_athlete_event_results_df,
    olympic_games_df.filter(items=['edition_id', 'year', 'season']),
    how='left',
    on='edition_id')

agg_athlete_event_results_df.info()
```

```python
filtered_agg_athlete_event_results_df = agg_athlete_event_results_df.filter(items=['year', 'country_noc', 'sex', 'season'])
athlete_to_year_df = filtered_agg_athlete_event_results_df.groupby(['year','sex', 'season'])['year'].count().reset_index(name="player_count")
athlete_to_year_df
```

```python
summer_athlete_to_year_df = athlete_to_year_df[athlete_to_year_df['season'] == 'Summer']
sns.lineplot(data=summer_athlete_to_year_df, x='year', y='player_count', hue='sex')
```

```python
winter_athlete_to_year_df = athlete_to_year_df[athlete_to_year_df['season'] == 'Winter']
sns.lineplot(data=winter_athlete_to_year_df, x='year', y='player_count', hue='sex')
```

# 4. Feature Engineering

```python
# Feature engineering - BMI
def isfloat(num):
    try:
        float(num)
        return True
    except ValueError:
        return False

# BMI is weight in kilograms divided by height in meters squared. BMI = kg/m2
# A BMI of 25.0 or more is overweight, while the healthy range is 18.5 to 24.9. BMI applies to most adults 18-65 years
def feature_bmi_creation(row):
    if isfloat(row['weight']) and isfloat(row['height']):
        return float(row['weight']) / (float(row['height']) / 100)**2
    else:
        return 'na'

agg_athlete_event_results_df['BMI'] = agg_athlete_event_results_df.apply(lambda row: feature_bmi_creation(row), axis=1)
```

```python
# Feature engineering - age_competing

def isdateYYYYMMDD(date):
    try:
        datetime.strptime(str(date),'%Y-%m-%d')
        return True
    except ValueError:
        return False

def feature_age_competing(row):
    if isdateYYYYMMDD(row['born']) and isdateYYYYMMDD(row['start_date']):
        delta_time = datetime.strptime(str(row['start_date']),'%Y-%m-%d') - datetime.strptime(str(row['born']),'%Y-%m-%d')
        return delta_time.days / 365.25
    else:
        return 'na'

agg_athlete_event_results_df['age_competing'] = agg_athlete_event_results_df.apply(lambda row: feature_age_competing(row), axis=1)
```
